## 3. Verify it with queries (grain, counts, missing values, windows)

*Every claim above gets a query cell here. A contract claim without a query next to it is a guess.*

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Basic information
print("=" * 50)
print("DATASET SUMMARY")
print("=" * 50)
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

# Missing values
print("\n" + "=" * 50)
print("MISSING VALUES")
print("=" * 50)
missing = df.isnull().sum()
print(missing[missing > 0].sort_values(ascending=False))

# Duplicate rows
print("\n" + "=" * 50)
print("DUPLICATE ROWS")
print("=" * 50)
print(df.duplicated().sum())

# Date range (only if a date column exists)
print("\n" + "=" * 50)
print("DATE INFORMATION")
print("=" * 50)

date_cols = [c for c in df.columns if "date" in c.lower()]

if date_cols:
    for col in date_cols:
        df[col] = pd.to_datetime(df[col], errors="coerce")
        print(f"{col}:")
        print("  Earliest:", df[col].min())
        print("  Latest  :", df[col].max())
else:
    print("No date column found in this dataset.")

# Show column names
print("\n" + "=" * 50)
print("COLUMN NAMES")
print("=" * 50)
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")
```

The data contract is verified by checking:

- dataset size
- missing values
- duplicate rows
- reporting date range

These checks confirm that the dataset matches the expected structure before feature engineering begins.

## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

### Data Limits

This dataset contains historical observations only.

It cannot prove that refreshing content causes higher rankings or more traffic.

Historical metrics may contain seasonal effects, incomplete tracking, or missing observations.

The model should be used for decision support rather than making guaranteed predictions.
