## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

### ML Task Type: Ranking

This is primarily a ranking task. The goal is to rank content pages by their priority for refresh so the content team can update the most valuable pages first. A ranking approach is more useful than a simple yes/no classification because resources are limited and pages must be ordered by expected value.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

### Target / Proxy

The target is a refresh priority score for each content page. Since no perfect ground-truth label exists, the model will learn from historical performance metrics such as impressions, clicks, pageviews, engagement, and content age to estimate which pages are better candidates for a refresh.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

### Success Metric

The primary success metric is Precision@K (for example, Precision@100). A good model should place the highest-value refresh candidates near the top of the ranked list so the content team spends its effort on pages with the greatest expected benefit.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

### Why Machine Learning?

A fixed rule such as "refresh pages older than one year" ignores many important factors. Refresh priority depends on multiple interacting signals, including impressions, clicks, engagement, traffic trends, and content age. Machine learning can combine these signals to produce a better ranking than a simple if-statement.
