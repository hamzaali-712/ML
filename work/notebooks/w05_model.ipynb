## 2. Split design

*Grouped by client? Time-aware? Say why this split is honest for your question.*

The dataset was split into training and testing sets using the same split strategy as the Week-4 baseline. The split prevents information leakage by ensuring the test data is never used during training. Using the same split allows a fair comparison between the baseline and the new model.

The data was split chronologically so that older records were used for training and newer records for testing. This reflects how the model would be used in production and avoids future information leaking into training.
