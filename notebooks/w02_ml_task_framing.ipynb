# 1. My Lane as an ML Task

## Selected Lane
Content Performance Prediction (FlyRank)

## ML Task Type
Binary Classification

## Problem Statement

The objective of this project is to predict whether a content page is likely to become a high-performing page based on the available features in the dataset.

The prediction helps prioritize which pages should receive SEO optimization, updates, or promotion.

This supports data-driven decision making instead of relying on manual judgment.

# 2. Target or Proxy

The dataset does not provide a perfect business outcome label.

Therefore, I will create a proxy target called **high_performance**.

The target will classify each content page into one of two categories:

- **1** = High-performing content
- **0** = Not high-performing

The exact definition will depend on the engagement or performance-related information available in the dataset.

# 3. Success Metric

This is a binary classification problem.

The model will be evaluated using:

- Accuracy
- Precision
- Recall
- F1 Score

The primary evaluation metric is **F1 Score** because it balances Precision and Recall and performs well when class distributions are uneven.

# 4. Unit of Analysis

Each row in the dataset represents one content page or content item.

Each row contains information describing that single content item, including its content characteristics and performance-related features.

Therefore,

**One row = One content page.**

# 5. Why Machine Learning Beats a Fixed Rule

A simple rule such as:

"If performance is greater than a certain threshold, then the page is successful."

cannot capture the complex relationships between multiple variables.

Content performance depends on many factors working together, including content characteristics, historical performance, and other available features.

Machine learning can learn these complex relationships from historical data and make more accurate predictions than manually designed rules.

# 6. Business Action

The prediction produced by the model supports real business decisions.

If a page is predicted to perform well:

- Prioritize promotion
- Invest additional SEO effort
- Share across marketing channels

If a page is predicted to perform poorly:

- Refresh the content
- Improve metadata
- Optimize keywords
- Update internal links

This allows the content team to focus resources where they are most likely to improve results.
