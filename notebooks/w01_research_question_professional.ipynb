## 2. Research Question

**Decision:** Which pages should the SEO team review first?

**Unit of Analysis:** One content page.

**Output:** Ranked review priority.

**Action:** Refresh, expand, rewrite, monitor, or protect pages.

**Cost of Wrong Recommendation:** Missing important pages wastes traffic opportunities; recommending poor pages wastes team effort.

**Why ML:** Multiple search and engagement signals interact in complex ways, making ML useful for prioritization.

## 3. Careful Words

- This notebook identifies candidates for review.
- It does not prove that refreshing a page causes higher traffic.
- Results are decision support for humans.
